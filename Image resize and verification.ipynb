```python
import cv2
import glob
import numpy as np

X_data = []
file = glob.glob (r"D:/IRDE/Samples/v4/v4a frames/*.jpg")
for myFile in file:
    print(myFile)
    image = cv2.imread (myFile)
    X_data.append (image)
print('X_data shape:', np.array(X_data).shape)

x_data = []
files = glob.glob (r"D:/IRDE/Samples/v4/v4b frames/*.jpg")
for myfile in files:
    print(myfile)
    image = cv2.imread (myfile)
    image = cv2.resize(image,(240,240))
    x_data.append (image)
print('x_data shape:', np.array(x_data).shape)


```
